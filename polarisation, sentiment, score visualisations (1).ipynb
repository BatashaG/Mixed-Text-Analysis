```python
import pandas as pd
import matplotlib.pyplot as plt


# Group the data by 'topic_title' and calculate the average 'score' and 'sentiment_score'
grouped_data = df.groupby('topic_title').agg({
    'score': 'mean',
    'sentiment_score': 'mean'
}).reset_index()

# Get the repeated words with the highest average sentiment scores
top_positive_sentiment_words = grouped_data.nlargest(10, 'sentiment_score')

# Get the repeated words with the most negative average sentiment scores
top_negative_sentiment_words = grouped_data.nsmallest(10, 'sentiment_score')

# Combine the two dataframes to get the top repeated words with both highest and most negative sentiment scores
top_sentiment_words = pd.concat([top_positive_sentiment_words, top_negative_sentiment_words])

# Create the figure and the axes for the bar plot
plt.figure(figsize=(12, 6))
plt.bar(top_sentiment_words['repeated_word'], top_sentiment_words['sentiment_score'],color='purple')
plt.xlabel('Topic')
plt.ylabel('Average Sentiment')
plt.title('Average Sentiment by Topic')
plt.xticks(rotation=90)
plt.savefig('sentiment.png', dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Group the data by 'repeated_word' and calculate mean and standard deviation of 'sentiment_score'
grouped_data = df.groupby('topic_title')['sentiment_score'].agg(['mean', 'std']).reset_index()

# Calculate the coefficient of variation (CV) for each topic
grouped_data['CV'] = (grouped_data['std'] / grouped_data['mean']) * 100

# Set the threshold value for high positive or high negative CV
threshold = 500

# Filter the grouped_data to include only labels with high positive or high negative CV
high_positive_cv = grouped_data[grouped_data['CV'] > threshold]
high_negative_cv = grouped_data[grouped_data['CV'] < -threshold]
selected_labels = pd.concat([high_positive_cv, high_negative_cv])

# Plot the coefficient of variation (CV) for the selected labels using a bar plot
plt.figure(figsize=(10, 6))
plt.bar(selected_labels['topic_title'], selected_labels['CV'])
plt.xlabel('Topics')
plt.ylabel('Coefficient of Variation (CV)')
plt.title('Coefficient of Variation (CV) for Selected Topics')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
